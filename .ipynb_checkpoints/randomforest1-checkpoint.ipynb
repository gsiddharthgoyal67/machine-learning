```python
import pandas as pd
```

```python
#loading datasets
df=pd.read_csv("social.csv")
df.info()
```

```python
#removing user id
filterdata=df.iloc[0:,2:].values
```

```python
#now we can apply training test
features=filterdata[0:,0:2]
```

```python
#label finding
label=filterdata[0:,2]
```

```python
from sklearn .model_selection import train_test_split
```

```python
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```python
#now apply fesrures scalling
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
#now apply scalling in training data only
X=sc.fit_transform(X)
```

```python
X
```

```python
x=sc.transform(x)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#random forest call with no f decision tree
Rclf=RandomForestClassifier(n_estimators=20)
```

```python
#now training
trained=Rclf.fit(X,Y)
```

```python
#now predicting
output=trained.predict(x)
```

```python
y
```

```python
from sklearn.metrics import accuracy_score
rate=(accuracy_score(output,y)*100)
rate
```
